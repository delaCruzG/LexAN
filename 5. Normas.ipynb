```python
import networkx as nx
grafo_csmx = nx.read_gexf("grafo_csmx_ventana10.gexf")
peso=nx.get_edge_attributes(grafo_csmx,'freq')
```

```python
import pandas as pd
```

```python
import os
import re
def limpiar_espacios(cadena):
    # Eliminar espacios después de paréntesis de apertura y antes de paréntesis de cierre
    cadena = re.sub(r'\(\s+', '(', cadena)  # Quita espacios después de '('
    cadena = re.sub(r'\s+\)', ')', cadena)  # Quita espacios antes de ')'
    # Eliminar espacios al inicio y al final de la cadena
    cadena = cadena.strip()
    return cadena
```

```python
terminos = {}
for archivo in os.listdir("terminos"):
    original = open("terminos/"+archivo)
    cadena_original = original.readline()
    traducido = open("terminos_lematizados/"+archivo)
    cadena_traducida = traducido.readline()
    print(cadena_original+":"+limpiar_espacios(cadena_traducida))
    terminos[cadena_original] = limpiar_espacios(cadena_traducida)
```

```python
terminos_lematizados = []
for llave in terminos.keys():
    terminos_lematizados.append(terminos[llave])
```

```python
def unir_multipalabras(lista_palabras):    
    return [palabra.replace(" ", "_") for palabra in lista_palabras]
```

```python
terminos_lematizados_unidos = unir_multipalabras(terminos_lematizados)
```
